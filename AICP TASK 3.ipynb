```python
# Function to calculate and display cost for slab 1
def costSlab1():
    unit_range = 100
    cost_per_unit = 10
    slab_data = electricity_data[0]
    total_cost = sum(min(unit, unit_range) * cost_per_unit for unit in slab_data)
    print(f"Cost for Slab 1: Rs. {total_cost}")
    return total_cost

# Function to calculate and display cost for slab 2
def costSlab2():
    if len(electricity_data) > 1:
        unit_range = 200
        cost_per_unit = 15
        slab_data = electricity_data[1]
        total_cost = sum((min(unit, unit_range) - 100) * cost_per_unit for unit in slab_data)
        print(f"Cost for Slab 2: Rs. {total_cost}")
        return total_cost
    else:
        print("Insufficient data for Slab 2")
        return 0

# Function to calculate and display cost for slab 3
def costSlab3():
    if len(electricity_data) > 2:
        unit_range = 300
        cost_per_unit = 20
        slab_data = electricity_data[2]
        total_cost = sum((min(unit, unit_range) - 200) * cost_per_unit for unit in slab_data)
        print(f"Cost for Slab 3: Rs. {total_cost}")
        return total_cost
    else:
        print("Insufficient data for Slab 3")
        return 0

# Get student ID as input
student_id = input("Enter your student ID: ")

# Main menu
while True:
    print("\nStudent ID:", student_id)
    print("1. Display bill for Slab 1 and Slab 2")
    print("2. Display bill for Slab 3")
    print("Press any other key to exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        total_cost_slab1 = costSlab1()
        total_cost_slab2 = costSlab2()
        total_cost_all = total_cost_slab1 + total_cost_slab2
        print(f"Total Cost for Slab 1 and Slab 2: Rs. {total_cost_all}")
    elif choice == "2":
        total_cost_slab3 = costSlab3()
        print(f"Total Cost for Slab 3: Rs. {total_cost_slab3}")
    else:
        break
```
